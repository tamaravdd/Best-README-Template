# 2. Where will I store my data and files?

**Option 1: Classic Local File Structure (“The Safe Default”)**

A clear file structure is essential for keeping your work organized.

We recommend the following structure:


```python
project-name/
│
├── data/
│   ├── raw/
│   └── cleaned/
│
├── notebooks/
│
├── outputs/
│   ├── figures/
│   └── tables/
│
├── posted/
│
└── README.md
```

Key ideas:

- raw data → never modify
- cleaned data → prepared for analysis
- notebooks → your analysis (with notes for different versions)
- outputs → plots and results
- posted → what you sent to collaborators/meeting (so you can easily go back to that version)
- README.md (A README file is a central document that explains what your project does, how it is structured, and how to navigate it. It is the entry point to any professional data or coding project.)

**High-level: The README file**

The README is the entry point to your project. If someone opens your work for the first time, they should read the README to understand what you did, why it matters, and how everything is organized.

In other words, it is your project in one page.

The README is

- a high-level explanation
- a guide to your project
- a summary of key decisions

The README is NOT

- a detailed log of everything you did
- a place for trial-and-error
- a replacement for your notebook


Why is this important?
In professional settings, you work is reviewed by others and often reused. It is therefore critical for your code and analysis to be communciated clearly.


Potential template

```python
# Project Title

## Objective
What business question are you answering?

## Data
What data are you using?

## Project Structure
- data/: datasets
- notebooks/: analysis
- outputs/: figures

## Approach
What did you do at a high level?

## Key Insights
- Insight 1
- Insight 2
- Insight 3

## Limitations
What are the main caveats?

## Next Steps
What would you do if you had more time?
```

- throughout → reasoning and decisions
